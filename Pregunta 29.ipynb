## Pregunta 29: Utilice los logs para crear una matriz cuyas columnas sean los logtypes, los indices las actions y las celdas la cantidad de la interseccion de ambas

```python
import pandas as pd
import numpy as np
```

```python
columns=['logtype','action']
logs_df=pd.read_csv("logs.csv",usecols=columns)
logs_df=logs_df.astype({'action':'category','logtype':'category'})
logs_df['interaccion']=1
logs_df=pd.pivot_table(logs_df,values=['interaccion'],index='logtype',columns='action',aggfunc='sum',fill_value=0)
logs_df.columns=logs_df.columns.droplevel()
```
