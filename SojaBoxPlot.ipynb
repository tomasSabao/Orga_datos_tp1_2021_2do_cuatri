```python
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib.pyplot import figure
import matplotlib.pyplot as plt
```

```python
columnas=['Fecha','Cierre']
soja_anio=pd.read_csv("datasetRofex4 - datasetRofex2.csv",usecols=columnas)
soja_anio['Fecha']=pd.to_datetime(soja_anio['Fecha'])
soja_anio['Año']=pd.DatetimeIndex(soja_anio['Fecha']).year
soja_anio=soja_anio.rename(columns={'Cierre': 'Precio_de_ajuste'})
```

```python
sns.set_theme(style="whitegrid")
fig, ax = plt.subplots(figsize=(10,8))
plt.title("Precio de ajuste de la soja por año")
sns.boxplot(x=soja_anio['Año'],y=soja_anio['Precio_de_ajuste'])
```
