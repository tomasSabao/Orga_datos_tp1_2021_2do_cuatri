## Pregunta 37 :Obtenga un dataframe que tenga como índice al user_id, como columnas a los idiomas y el nivel de cada usuario para cada idioma como valor con -1 en caso de no tenerlo cargado

```python
idioma_df=pd.read_csv("languages.csv")
idioma_df=idioma_df.dropna()
idioma_df=idioma_df.astype({'babel_lang':'category'})
idioma_df=idioma_df.pivot(index='babel_user',columns='babel_lang',values='babel_level')
idioma_df=idioma_df.fillna(-1)
```
